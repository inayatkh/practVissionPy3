# Detecting faces in images

- We will write a simple Python + OpenCV 3 code to detect the presence of a face in an image
- We will use this image
<img src=basit.jpg/>

```python
# import the required packages
import cv2
cv2.__version__
```

```python

# load the image and convert it to gray scale
image = cv2.imread("basit.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.equalizeHist(gray)

# Open a new thread to manage the external cv2 interaction
#cv2.startWindowThread()

```

```python

# load the haar cascade face detector
detector = cv2.CascadeClassifier("./haarcascade_frontalface_alt.xml")
rects = detector.detectMultiScale(gray, scaleFactor=1.05, minNeighbors=7,
                                 minSize=(30,30), flags=cv2.CASCADE_SCALE_IMAGE)

# loop over the faces and draw a rectangle surrounding each face
for (x, y, w, h) in rects:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255,0), 2)
```

```python

# Open a new thread to manage the external cv2 interaction
#cv2.startWindowThread()
cv2.imshow("Faces", image)
#cv2.namedWindow("Faces", cv2.WINDOW_NORMAL)
# show the image
cv2.imshow("Faces", image)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```
